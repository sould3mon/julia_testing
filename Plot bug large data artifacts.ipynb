```julia
using Dates
using DataFrames
file = "GBPJPY1.csv" #65k minute Data points 
my_names = ["Date","Time","open","high","Low", "close","Volume"]


f_open = readcsv(file)

#date_fmt = DateFormat("y.m.d")
#time_fmt = DateFormat("H:M")
#data_format = [string,string,float,float,float,float,int]


df = DataFrame(f_open)

names!(df.colindex,map(parse,my_names))
#tempval my_start my_end to show
my_start = 2000 #from the end
my_end = 1 #variable from the end to watch overlaping set
using Immerse

# only plotting 1 dataset for y normaly plot open high low and close) 

Gadfly.set_default_plot_size(18cm, 11cm)
Theme(minor_label_font_size=6pt)
data_size=df[:Time][end-my_start:end-my_end]
plot(x=data_size,y=df[:close][end-my_start:end-my_end],Geom.line,
Guide.xlabel("time"),
Guide.xticks(ticks=[1:30:length(data_size)],orientation=:vertical)
)

```
